```python
import os
import csv
import pandas as pd
import numpy as np
```

```python
game_file = os.pardir + '/input_data/kaggle_data/kaggle_games.csv'
at_bats_file = os.pardir + '/input_data/kaggle_data/kaggle_at_bats.csv'
pitches_file = os.pardir + '/input_data/kaggle_data/kaggle_pitches.csv'
player_file = os.pardir + '/input_data/kaggle_data/kaggle_player_names.csv'

output_file = os.pardir + '/output_data/merged_kaggle_data.csv'
```

```python
games_df = pd.read_csv(game_file)
at_bats_df = pd.read_csv(at_bats_file)
pitches_df = pd.read_csv(pitches_file)
players_df = pd.read_csv(player_file)
```

```python
games_df['date'] = (games_df['date'] + ' ' + games_df['start_time']).apply(pd.to_datetime)
games_df.rename(columns={'date':'date_time'}, inplace=True)
games_df = games_df[games_df['date_time'].dt.year == 2018]
games_df = games_df[['away_team', 'date_time', 'g_id', 'home_team']]
```

```python
games_df['away_team'] = games_df['away_team'].apply(str.upper)
games_df['home_team'] = games_df['home_team'].apply(str.upper)
```

```python
games_df = games_df[['g_id', 'home_team', 'away_team', 'date_time']]

games_data_types = {'g_id': str}
games_df = games_df.astype(games_data_types)
```

```python
at_bats_df = at_bats_df[['ab_id', 'g_id', 'pitcher_id', 'batter_id', 'inning', 'top', 'p_score', 'p_throws', 'stand']]

at_bats_data_types = {'ab_id':str, 'g_id':str, 'pitcher_id':str, 'batter_id':str}
at_bats_df = at_bats_df.astype(at_bats_data_types)
```

```python
at_bats_df = at_bats_df[at_bats_df['g_id'].str[0:4] == '2018']
```

```python
pitches_data_types = {'ab_id':str}
pitches_df = pitches_df.astype(pitches_data_types)

pitches_df = pitches_df[pitches_df['ab_id'].str[0:4] == '2018']
pitches_df['ab_id'] = pitches_df['ab_id'].str[:-2]
```

```python
players_df['name'] = players_df['first_name'] + ' ' + players_df['last_name']
players_data_types = {'id':str}
players_df = players_df.astype(players_data_types)
```

```python
print(players_df.nunique())
```

```python
players_df[players_df.duplicated(subset=['name'], keep=False)]
# problem: with the kaggle dataset, no way to tell duplicate players apart
# maybe can use context clues - pitchers vs. batters, what team the player is on, etc.
# maybe just hope none of these players is involved in a stolen base attempt
```

```python
new_df = pd.merge(at_bats_df, pitches_df, on=['ab_id'])
new_df = pd.merge(games_df, new_df, on=['g_id'])
```

```python
pitcher_names = [None] * len(new_df)
batter_names = [None] * len(new_df)
for i in range(len(new_df)):
    pitcher_index = players_df[players_df['id'] == new_df['pitcher_id'][i]].index[0]
    batter_index = players_df[players_df['id'] == new_df['batter_id'][i]].index[0]
    pitcher_names[i] = players_df.iloc[pitcher_index]['name']
    batter_names[i] = players_df.iloc[batter_index]['name']
```

```python
new_df['pitcher_name'] = pitcher_names
new_df['batter_name'] = batter_names
```

```python
new_df.to_csv(output_file, index=False)
```
